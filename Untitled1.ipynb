```python
import os
import math
import numpy as np
from src.data import create_blueprint, sample_distribution_parameters, add_gaussian_noise, normalize, generate_outliers
from src.data import generate_from_blueprint
from src.outliers import generate_subspace_outliers, generate_contextual_outliers

import matplotlib as mpl
from matplotlib import gridspec
import matplotlib.pyplot as plt
import seaborn as sns
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{libertine}'
mpl.rc('font', family='serif')
#sns.set()
```

```python
num_devices = 5
num_data = 100
dims = 2
dims_outliers = 1
frac_outliers_devices = 0.1
```

```python
bp = create_blueprint(dims)
params = sample_distribution_parameters(num_devices)
blueprints = np.array([add_gaussian_noise(bp, params=params[i], snr=0.2) for i in range(num_devices)])
outliers = np.array([generate_outliers(dbp, dims_outliers) for dbp in blueprints])
for i in range(len(outliers)):
    if np.random.uniform() > frac_outliers_devices:
        outliers[i].fill(-1)
```

```python
# generate data from blueprint
data = np.array([
    generate_from_blueprint(blueprints[i], num_data, snr=0.2) for i in range(num_devices)
])

# normalize
data = normalize(data)

# add outliers
outliers = np.expand_dims(outliers, axis=1)
outliers = np.repeat(outliers, num_data, axis=1)
is_outlier = np.not_equal(outliers, -1)
print(is_outlier.shape)
data[is_outlier] = outliers[is_outlier]    
```
