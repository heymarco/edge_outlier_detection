```python
from src.pipelines import dataset_drift
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
num_devices = 3
dims = 2
T_start = 1
T = 10
n = 4
```

```python
data = dataset_drift(num_devices, n*(T+T_start), dims, 20, frac_drifting_devices=0.5)
```

```python
def plot_data(data):
    if dims < 3:
        for d in data:
            plt.scatter(d.T[0], d.T[1])
    else:
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        for device_data in data:
            x = device_data.T[0]
            y = device_data.T[1]
            z = device_data.T[2]
            ax.scatter(x, y, z)
        plt.show()
```

```python
plot_data(data)
```
